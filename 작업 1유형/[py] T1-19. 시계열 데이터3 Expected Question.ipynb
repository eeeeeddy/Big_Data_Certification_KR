### 주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과 
### 2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오. 
### 단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함 
### (최종값은 소수점 반올림 후 정수 출력)
- 데이터셋 : basic2.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic2.csv')
df.head()
```

```python
# Date 데이터 타입 변경
df.Date = pd.to_datetime(df.Date)

# 확인
df.info()
```

```python
# Year, Month 컬럼 추가

df['Year'] = df.Date.dt.year
df['Month'] = df.Date.dt.month
df
```

```python
# Events가 1일 경우 80%의 Sales 값으로 변경

def changeSale(x):
    if x['Events'] == 1:
        x['Sales'] *= 0.8
    else:
        x['Sales'] = x['Sales']
    return x

df = df.apply(lambda x : changeSale(x), axis=1)
df
```

```python
# 2022년 월별 Sales 합계 중 가장 큰 금액 

max2022 = df[df.Year == 2022].groupby('Month').sum().Sales.max()
max2022
```

```python
# 2023년 월별 Sales 합계 중 가장 큰 금액 (Events가 1일 경우 80%의 Sales 값 반영)

max2023 = df[df.Year == 2023].groupby('Month').sum().Sales.max()
max2023
```

```python
# 두 값의 차이를 절대값으로, 반올림하여 정수 출력

int(round(abs(max2022 - max2023), 0))
```
