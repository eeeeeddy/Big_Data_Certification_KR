### 2022년 5월 sales의 중앙값을 구하시오

- 데이터셋 : basic2.csv 
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic2.csv')
df
```

```python
# Date의 데이터 타입 확인

df.Date.dtype
```

```python
# Date의 데이터 타입을 datetime으로 변경
df.Date = pd.to_datetime(df.Date)
df.Date.dtype
```

```python
# 문제를 풀기 위해 Year와 Month를 따로 구하여 새롭게 column 생성
df['Year'] = df.Date.dt.year
df['Month'] = df.Date.dt.month
```

```python
# 2022년 5월 Sales의 중앙값 구하기
df[(df.Year == 2022) & (df.Month == 5)].Sales.median()
```
