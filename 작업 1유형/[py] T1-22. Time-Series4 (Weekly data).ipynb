### 주어진 데이터(basic2.csv)에서 주 단위 Sales의 합계를 구하고, 가장 큰 값을 가진 주와 작은 값을 가진 주의 차이를 구하시오(절대값)
- 데이터셋 : basic2.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

````python
import pandas as pd

df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic2.csv')
df.head()
````

````python
# Date의 데이터 타입을 datetime으로 변경

df.Date = pd.to_datetime(df.Date)
````

````python
# 인덱스를 Date로 설정

df = df.set_index(['Date'])
df
````

````python
# 데이터를 주 단위로 resample

df_w = df.resample('W').sum()
df_w
````

````python
abs(df_w.Sales.max() - df_w.Sales.min())
````
