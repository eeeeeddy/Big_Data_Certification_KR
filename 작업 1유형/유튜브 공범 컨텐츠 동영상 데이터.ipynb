14. 각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
#pd.set_option('display.max_rows',None)
pd.options.display.max_rows = 60
```

```python
ww = channel[(channel.ct >= '2021-10-03 03:00:00') & (channel.ct <= '2021-11-01 15:00:00')]
```

```python
wf = ww.drop_duplicates('channelname', keep='first')[['channelname','subcnt']].reset_index(drop=True)
wf
```

```python
wl = ww.drop_duplicates('channelname', keep='last')[['channelname','subcnt']].reset_index(drop=True)
wl
```

```python
wl.columns = ['channelname', 'wl_sub']
wf.columns = ['channelname', 'wf_sub']
```

```python
ans = pd.merge(wf, wl)
ans
```

```python
ans['del'] = ans.wl_sub - ans.wf_sub
ans[['channelname', 'del']]
```

16. 각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터 프레임으로 만들고 출력하라

```python
ww = video[['ct', 'videoname']].sort_values(['ct']).drop_duplicates('videoname', keep='first')
ww['date'] = ww['ct'].dt.date
ww[['date', 'videoname']]
```

17. “공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다. 공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라

```python
ww = video.sort_values(['ct','videoname'])[['videoname', 'viewcnt', 'ct']]
ww.loc[(ww.ct.dt.hour == 21) & (ww.ct.dt.minute == 0)].drop_duplicates('videoname', keep='first').sort_values('viewcnt', ascending=False).reset_index(drop=True)
```

18. video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라

```python
video['ratio'] = video['dislikecnt'] / video['likecnt']
ww = video[['videoname', 'ratio', 'ct']].sort_values('ct', ascending=False).drop_duplicates('videoname')
ww[['videoname', 'ratio']].sort_values('ratio').reset_index(drop=True)
```

19. 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
ww = video.loc[(video.ct.dt.month == 11) & (video.ct.dt.day == 1)]
ww = ww.loc[(ww.ct.dt.hour >= 0) & (ww.ct.dt.hour < 15)]
ww = ww[['viewcnt', 'ct', 'videoname']]

def delta(x):
    return max(x) - min(x)

ww[['viewcnt', 'videoname']].groupby('videoname').agg(delta)
```
