### 주어진 데이터(basic2.csv)에서 "pv"컬럼으로 1일 시차(lag)가 있는 새로운 컬럼을 만들고(예: 1월 2일에는 1월 1일 pv데이터를 넣고, 1월 3일에는 1월 2일 pv데이터를 넣음),새로운 컬럼의 1월 1일은 다음날(1월 2일)데이터로 결측치를 채운 다음, Events가 1이면서 Sales가 1000000이하인 조건에 맞는 새로운 컬럼 합을 구하시오
- 데이터셋 : basic2.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic2.csv')
df.head()
```

```python
df.Date = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df = df.set_index(['Date'])
df
```

```python
# pv값을 가진 시차가 있는 새로운 컬럼 만들기
df.PV.shift(periods=1, freq='D')[:-1]
```

```python
# 새로운 컬럼 df에 추가하기

df['lag_PV'] = df.PV.shift(periods=1, freq='D')[:-1]
df
```

```python
# 1월 1일 결측치 채우기 (1월 2일 데이터로)
df.lag_PV[0] = df.lag_PV[1]
df
```

```python
# Events가 1이면서 Sales가 1000000이하인 조건에 맞는 새로운 컬럼 합 구하기

df[(df.Events == 1) & (df.Sales <= 1000000)].lag_PV.sum()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv")
df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
#1일 차이가 나는 시차 특성 만들기
df['previous_PV'] = df['PV'].shift(1)
df.head()
```

```python
# 1일 씩 미뤘음으로 가장 앞이 결측값이 됨 (바로 뒤의 값으로 채움)
df['previous_PV'] = df['previous_PV'].fillna(method = 'bfill')
df.head()
```

```python
# 조건에 맞는 1일 이전 PV의 합
cond = (df['Events'] == 1) & (df['Sales'] <= 1000000)
print(df[cond]['previous_PV'].sum())
```
