# T1-29. 12월인 데이터 수는?

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/payment.csv")
df
```

```python
# your code

# date 컬럼을 활용하여 month 컬럼 만들기
df['month'] = df.date.astype('str').str[4:6]

# 12월의 데이터만 추출
df1 = df[df.month == '12']

len(df1)
```

```python
# 풀이
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
print(sum(df['date'].dt.month == 12))
```
