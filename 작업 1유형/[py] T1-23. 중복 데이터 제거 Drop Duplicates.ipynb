## f1의 결측치를 채운 후 age 컬럼의 중복 제거 전과 후의 'f1' 중앙값 차이를 구하시오 
### - 결측치는 f1의 데이터 중 내림차순 정렬 후 10번째 값으로 채움
### - 중복 데이터 발생시 뒤에 나오는 데이터를 삭제함
### - 최종 결과값은 절대값으로 출력

- 데이터셋 : basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic1.csv')
df
```

```python
# f1의 결측치 확인
df.f1.isnull().sum()
```

```python
# f1의 데이터 중 내림차순 정렬 후 10번째 값으로 결측치 채우기

val = df.f1.sort_values(ignore_index=True, ascending=False)[9]

df.f1 = df.f1.fillna(val)

val, df
```

```python
# 결측치가 잘 채워졌는 지 확인

df.f1.isnull().sum()
```

```python
# age 컬럼의 중복 제거 전 'f1' 중앙값 구하기
before = df.f1.median()
before
```

```python
# age 컬럼 중복 제거 (중복 데이터 발생시 뒤에 나오는 데이터를 삭제)

df1 = df.drop_duplicates(subset=['age'])
df1
```

```python
# age 컬럼의 중복 제거 후 'f1' 중앙값 구하기

after = df1.f1.median()
after
```

```python
# 두 중앙값의 차이 구하기 (절대값)

abs(before - after)
```
