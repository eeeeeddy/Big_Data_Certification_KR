# T1-30. 12월 25일 결제 금액(price)은 12월 총 결제금액의 몇 %인가? (정수로 출력)

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/payment.csv")
df
```

```python
# your code

# date 컬럼을 활용하여 month 컬럼 생성
df['month'] = df.date.astype('str').str[4:6]
df
```

```python
# 12월 총 결제금액 구하기

total = df[df.month == '12'].price.sum()
total
```

```python
# 12월 25일 결제 금액(price) 구하기

val1 = df[df.date == 20221225].price.sum()
val1
```

```python
# 12월 25일 결제 금액(price)은 12월 총 결제금액의 몇 %인가? (정수로 출력)

int(round(val1/total*100, 0))
```

```python
# 풀이
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
cond1 = df['date'].dt.month == 12
cond2 = df['date'].dt.day == 25
result = sum(df[cond1 & cond2]['price']) / sum(df[cond1]['price'])
print(int(result*100))
```
