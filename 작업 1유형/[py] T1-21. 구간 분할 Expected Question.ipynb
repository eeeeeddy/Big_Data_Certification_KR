## 나이 구간 나누기
###  basic1 데이터 중 'age'컬럼 이상치를 제거하고, 동일한 개수로 나이 순으로 3그룹으로 나눈 뒤 각 그룹의 중앙값을 더하시오
### (이상치는 음수(0포함), 소수점 값)
- data: basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작


```python
import pandas as pd

df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic1.csv')
df.head()
```

```python
# age가 음수인 데이터들의 인덱스 찾기

idx1 = df[df.age <= 0].index.to_list()
idx1
```

```python
# age가 float 타입인 데이터들의 인덱스 찾기

idx2 = df[(df.age % 1 >= 0.1) & (df.age % 1 <= 0.9)].index.to_list()
idx2
```

```python
# age가 음수, float 타입인 인덱스 찾기 (idx1, idx2 합집합)

outlier_idx = list(set(idx1) | set(idx2))
list(outlier_idx)
```

```python
# 이상치 제거

df = df.drop(labels=outlier_idx)
```

```python
df.age.unique()
```

```python
df = df.sort_values('age', ascending=False, ignore_index=True)
```

```python
# 같은 개수로 나누어 각 중앙값의 합 구하기

df.age[:30].median() + df.age[30:60].median() + df.age[60:].median()
```

### 풀이

```python
import pandas as pd
```

```python
df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
```

```python
# age 이상치 (음수(0포함), 소수점 제거)
print('전체 데이터:', df.shape)
df = df[~(df['age'] <= 0)]
print('음수(0포함)값 제거 후 데이터 크기:', df.shape)

df = df[(df['age'] == round(df['age'],0))]
print('소수점 제거 후 데이터 크기:', df.shape)
```

```python
df.head()
```

```python
# 기준 확인
pd.qcut(df['age'], q=3)
```

```python
# 구간 분할
df['range'] = pd.qcut(df['age'], q=3, labels=['group1','group2','group3'])
```

```python
# 수량 비교
df['range'].value_counts()
```

```python
# 중간이상 - 중간이하 
g1_med = df[df['range'] == 'group1']['age'].median()
g2_med = df[df['range'] == 'group2']['age'].median()
g3_med = df[df['range'] == 'group3']['age'].median()

print(g1_med + g2_med + g3_med)
```
