## 조건에 맞는 데이터 표준편차 구하기
- 주어진 데이터 중 basic1.csv에서 'f4'컬럼 값이 'ENFJ'와 'INFP'인 'f1'의 표준편차 차이를 절대값으로 구하시오
- 데이터셋 : basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic1.csv')
df
```

```python
# 조건에 맞는 데이터 (ENFJ, INFP)
df = df[(df.f4 == 'ENFJ') | (df.f4 == 'INFP')]
df
```

```python
# 조건에 맞는 f1의 표준편차 (ENFJ, INFP)
print("ENFJ:", df[df.f4=='ENFJ'].f1.std(), "INFP:", df[df.f4=='INFP'].f1.std())
```

```python
# 두 표준편차 차이 절대값 출력
np.abs(df[df.f4=='ENFJ'].f1.std()-df[df.f4=='INFP'].f1.std())
```

## Hint
- 절대 값 : np.abs()
- 표준편차 : DataFrame.std()
