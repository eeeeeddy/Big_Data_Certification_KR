상하위 5% 값의 합을 구할 때, 상하위 5% 위치에 있는 모든 값들의 합으로 생각하여 다음과 같이 구하였었다.

```py
df[(df['f5'] >= 0.95) | (df['f5'] <= 0.05)].sum()
```

풀이를 보니 상하위 5%의 '값'의 합이었다. (quantile을 이용해야함을 의미)

````python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('/kaggle/input/bigdatacertificationkr/basic1.csv')
df
````

````python
# EDA
df.f5.isnull().sum()
````

````python
df.f5
````

````python
# min-max scale 
scaler = MinMaxScaler()
df.f5 = scaler.fit_transform(df[['f5']])
df.f5
````

````python
df.f5.quantile(0.95) + df.f5.quantile(0.05)
````
