## 주어진 데이터는 고혈압 환자 치료 전후의 혈압이다. 해당 치료가 효과가 있는지 대응(쌍체)표본 t-검정을 진행하시오
- 귀무가설(H0): $\mu$ >= 0
- 대립가설(H1): $\mu$ < 0
- $\mu$ = (치료 후 혈압 - 치료 전 혈압)의 평균
- 유의수준: 0.05

1. $\mu$의 표본평균은?(소수 둘째자리까지 반올림)
2. 검정통계량 값은?(소수 넷째자리까지 반올림)
3. p-값은?(소수 넷째자리까지 반올림)
4. 가설검정의 결과는? (유의수준 5%)

```python
import pandas as pd

df = pd.read_csv("/kaggle/input/bigdatacertificationkr/high_blood_pressure.csv")

# your code
# (1)
df['diff'] = df['bp_post'] - df['bp_pre']
print(round(df['diff'].mean(), 2))

# (2)
from scipy import stats
# alternative 선택 방법
# "후" < "전" : less, "후" > "전": greater
#  치료 후의 혈압이 떨어져야하므로("후" < "전") 'less' 선택
result = stats.ttest_rel(df['bp_post'], df['bp_pre'], alternative='less')
# st, pv = stats.ttest_rel(df['bp_post'], df['bp_pre'], alternative='less')
print(result)
print("검정통계량 :", round(result.statistic, 4))

# (3)
print("p-value :", round(result.pvalue, 4))

# (4)
# p-value가 0.05보다 작으므로 귀무가설을 기각하고 대립가설을 채택한다.
```

```python
import pandas as pd
from scipy import stats
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/high_blood_pressure.csv")

df['diff'] = df['bp_post'] - df['bp_pre']

#1
print(round(df['diff'].mean(),2))

#2 
st, pv = stats.ttest_rel(df['bp_post'], df['bp_pre'], alternative="less")
print(round(st,4))

#3
print(round(pv,4))

#4 귀무가설 기각, 대립가설 채택
```
