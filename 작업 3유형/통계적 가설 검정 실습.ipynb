##### 1번 단일표본 T-검정 실습
주어진 데이터에는 블랙 체리나무 31그로의 둘레와 높이, 부피가 저장되어 있다. <br>
이 표본의 평균이 모평균과 일치하는지 단일표본 t-검정을 통해 답하고자 한다. 가설은 아래와 같다

-  $H_0 :  \overline{X} = 75$
-  $H_1 : \overline{X}  \neq  75$
   -  Girth : 둘레
   -  Height : 높이
   -  Volume : 부피
단, 데이터의 각 변수들은 정규 분포를 만족한다고 가정한다.

```python
import pandas as pd
df = pd.read_csv('./trees.csv')
df
```

```python
from scipy import stats

# 표본평균을 구하시오
print(df.Height.mean())

# 위의 가설을 검정하기 위한 검정통계량을 구하시오 (반올림하여 소수 둘째자리까지 계산)
st, pv = stats.ttest_1samp(df.Height, 75)
print(round(st, 2))

# 위의 통계량에 대한 p-value값을 구하고(반올림하여 소수 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오
print(round(pv, 4))
if pv < 0.05:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```

##### 2번 독립표본 T-검정 실습
주어진 데이터에는 기니피그 60마리의 치아 길이, 투여한 비타민 C 종류, 투여량이 저장되어 있다. <br>
오렌지주스를 투여받은 기니피그의 치아 길이 평균이 아스코르브산을 투여받은 기니피그의 치아 길이 평균과 일치하는지 독립표본 T-검정을 통해 답하고자 한다. <br>
가설은 아래와 같다.

-  $H_0 :  \overline{X_{OJ}} = \overline{X_{VC}}$
-  $H_1 : \overline{X_{OJ}}  \neq \overline{X_{VC}}$
   -  len : 치아길이
   -  supp : 투여한 약제(VC: 아스코르브산, OJ: 오렌지주스)
   -  dose : 투여량
단, 데이터의 각 변수들은 정규분포를 만족하며, 두 그룹은 등분산을 띤다.

```python
import pandas as pd
df = pd.read_csv('./toothgrowth.csv')
df
```

```python
from scipy import stats

# 표본평균 X_OJ, X_VC를 구하시오 (반올림 소수점 둘째자리까지 계산)
print("오렌지주스 표본평균: ", round(df[df['supp']=='OJ'].len.mean(), 2))
print("아스코르브산 표본평균: ", round(df[df['supp']=='VC'].len.mean(), 2))

# 위의 가설을 검정하기 위한 검정통계량을 구하시오 (반올림 소수점 둘째자리까지 계산)
oj = df[df.supp == 'OJ'].len
vc = df[df.supp == 'VC'].len

st, pv = stats.ttest_ind(vc, oj, equal_var=True)
print(round(st, 2))

# 위의 통계량에 대한 p-value를 구하고(반올림 소수점 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오
print(round(pv, 2))
if pv < 0.05:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```

##### 3번 쌍체표본 T-검정 실습
주어진 데이터에는 여러 지역에서 스프레이를 뿌리기 전에 측정한 곤충의 수와 스프레이를 뿌린 다음의 곤충의 수가 저장되어 있다. <br>
해당 스프레이가 구제 효과가 있는지(즉, 스프레이를 뿌린 뒤 곤충의 수가 감소했는지) 쌍체표본 T-검정을 통해 답하고자 한다. <br>
가설은 다음과 같다.

-  $ \overline{\mu_d}$ : (살충제를 뿌린 후의 곤충 수 - 살충제를 뿌리기 전의 곤충 수)의 평균
-  $H_0 : \overline{\mu_d} \geq 0$
-  $H_1 : \overline{\mu_d} < 0$
   -  before_spr : 살충제를 뿌리기 전 곤충의 수
   -  after_spr : 살충제를 뿌린 뒤 곤충의 수
  
단, (살충제를 뿌린 후의 곤충 수 - 살충제를 뿌리기 전의 곤충 수)는 정규성을 만족한다.

```python
import pandas as pd
df = pd.read_csv('./insectsprays.csv')
df
```

```python
from scipy import stats

# (살충제를 뿌린 후의 곤충 수 - 살충제를 뿌리기 전의 곤충 수)의 평균의 표본평균을 구하시오 (반올림 소수 둘째자리까지 계산)
print((df['after_spr'] - df['before_spr']).mean())

# 위의 가설을 검정하기 위한 검정통계량을 구하시오 (반올림하여 소수 넷째자리까지 계산)
st, pv = stats.ttest_rel(df['after_spr'], df['before_spr'], alternative='less')
print(round(st, 4))

# 위의 통계량에 대한 p-value값을 구하고(반올림하여 소수 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오.
print(round(pv, 4))
if pv < 0.05:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```

##### 4번 일원분산분석
주어진 데이터는 붓꽃의 종, 꽃받침 길이와 너비, 꽃잎의 길이와 너비를 기록한 데이터이다. <br>
이를 이용하여 붓꽃의 종에 따라 꽃받침의 너비가 차이나는지 일원분산분석을 통해 답하고자 한다.<br>
가설은 다음과 같다.

-  $H_0 : \overline{X_1} = \overline{X_2} = \overline{X_3}$
-  $H_1 : {} \sim H_0 (H_0$가 아니다.)
  
단, 주어진 데이터가 정규성과 등분산성을 만족한다고 가정한다.

```python
import pandas as pd
df = pd.read_csv('./iris.csv')
df
```

```python
df.species.unique()
```

```python
from scipy import stats

# versicolor 종의 평균 꽃받침 너비를 구하시오. (반올림하여 소수 둘째자리까지 계산)
print(round(df[df['species']=='versicolor'].sepal_width.mean(), 2))

# 위의 가설을 검정하기 위한 검정통계량을 구하시오. (반올림하여 소수 둘째자리까지 계산)
# 분산분석을 진행하기 위하여 붓꽃 종별 꽃받침 너비 값들을 각각 X1, X2, X3에 저장 후 진행 (인덱스가 모두 같아야함)
X1 = df[df['species']=='setosa']['sepal_width']
X2 = df[df['species']=='versicolor']['sepal_width']
X3 = df[df['species']=='virginica']['sepal_width']

st = stats.f_oneway(X1, X2, X3)
print(round(st.statistic, 2))

# 위의 통계량에 대한 p-value값을 구하고(반올림하여 소수 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오.
print(round(st.pvalue, 4))
if st.pvalue < 0.04:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```

##### 5번 카이제곱검정(적합도 검정) 실습 ★
주어진 데이터에서 A 회사의 직원 200명을 대상으로 현재 사용하고 있는 스마트폰 제조사를 조사한 결과이다. <br>
그리고 최근 조사에 따르면 국내 스마트폰 제조사 점유율은 삼성 60%, 애플 30%, 기타 10%라고 한다. <br>
이를 이용하여 A 회사의 직원들이 사용하는 스마트폰 제조사 분포가 국내 스마트폰 제조사 점유율 분포와 동일한지 카이제곱검정을 통해 답하고자 한다. <br>
가설은 다음과 같다. 

-  $H_0 : $ A 회사 직원들이 사용하는 스마트폰 제조사 분포는 국내 스마트폰 제조사 점유을 분포를 따른다.
-  $H_1 : $ A 회사 직원들이 사용하는 스마트폰 제조사 분포는 국내 스마트폰 제조사 점유을 분포를 따르지 않는다.

```python
import pandas as pd
df = pd.read_csv('./cellphone.csv')
df
```

```python
from scipy import stats

count_df = df.groupby('제조사').count()
print(count_df)

# 삼성 제조사 스마트폰 사용 직원 수에서 애플 제조사 스마트폰 사용 직원 수를 뺀 값을 정수로 출력하시오.
print(len(df[df['제조사']=='삼성']) - len(df[df['제조사']=='애플']))

# 문제에서 제시된 국내 스마트폰 제조사 점유율을 이용하여 기대 빈도를 계산하고, 위의 가설을 검정하기 위한 검정통계량을 구하시오. (반올림 소수 둘째자리까지 계산)
total = len(df)
expected = [int(total*0.1), # '기타' 제조사 기대빈도
            int(total*0.6), # '삼성' 제조사 기대빈도
            int(total*0.3)] # '애플' 제조사 기대빈도 >> count_df 순서에 맞춰야함
print(expected) # 기대빈도

result = stats.chisquare(count_df['id'].values, f_exp=expected)
print(round(result.statistic, 2))

# 위의 통계량에 대한 p-value를 구하고(반올림 소수 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오.
print(round(result.pvalue, 4))
if result.pvalue < 0.05:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```

##### 6번 카이제곱검정(독립성 검정) 실습 ★
주어진 데이터는 A 대학의 학생 300명을 대상으로 각 학생이 속한 단과대학과 이번 학기 수강 과목 수를 조사한 결과이다. <br>
이를 이용하여 단과대학별로 수강 과목 수에 차이가 있는지 카이제곱검정을 통해 답하고자 한다.<br>
가설은 다음과 같다. 

-  $H_0 : $ 단과대학과 수강 과목 수는 관련이 없다.
-  $H_1 : $ 단과대학과 수강 과목 수는 관련이 있다.

```python
import pandas as pd
df = pd.read_csv('./student.csv')
df
```

```python
from scipy import stats

# 주어진 데이터로 단과대학과 수강 과목 수로 분할표를 만들었을 때, 수강 과목 수가 '6개 이상'인 학생 수가 가장 많은 단과대학을 출력하시오
table = pd.crosstab(df['단과대학'], df['수강 과목 수'])
print(table)
print(table[table['6개 이상']==table['6개 이상'].max()].index[0])

# 위의 가설을 검정하기 위한 검정통계량을 구하시오. (반올림 소수 둘째자리까지 계산)
X1 = table.loc['경영대학', :]
X2 = table.loc['자연대학', :]
X3 = table.loc['공과대학', :]

result = stats.chi2_contingency([X1, X2, X3])
print(round(result.statistic, 2))

# 위의 통계량에 대한 p-value 구하고(반올림 소수 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오.
print(round(result.pvalue, 4))
if result.pvalue < 0.05:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```

##### 마무리 문제 2번
주어진 데이터에는 4가지 종류의 식단으로 구분된 병아리들의 무게를 측정한 데이터가 저장되어 있다. <br>
이 중 1번 식단을 먹은 병아리의 평균 무게가 4번 식단을 먹은 병아리의 평균 무게와 일치하는지 독립표본 t-검정을 통해 답하고자 한다. <br>
가설은 아래와 같다.

-  $H_0 :  \overline{X_1} = \overline{X_4}$
-  $H_1 : \overline{X}  \neq \overline{X_4}$
   -  weight : 병아리의 무게
   -  Diet : 식단의 종류

단, 데이터의 weight 변수들은 정규 분포를 만족하며, 두 그룹은 등분산을 띠지 않는다.

```python
import pandas as pd
df = pd.read_csv('./chickweight.csv')
df
```

```python
from scipy import stats

# 1번 식단, 4번 식단에 대한 표본평균을 구하시오 (반올림하여 소수 둘째자리까지 계산)
samp1 = df[df['Diet'] == 1].weight
samp4 = df[df['Diet'] == 4].weight
print("1번 식단에 대한 표본 평균: ", round(samp1.mean(), 2))
print("4번 식단에 대한 표본 평균: ", round(samp4.mean(), 2))

# 위의 가설을 검정하기 위한 검정통계량을 구하시오 (반올림하여 소수 둘째자리까지 계산)
st, pv = stats.ttest_ind(samp1, samp4, equal_var = False)
print(round(st, 2))

# 위의 통계량에 대한 p-value 구하고(반올림 소수 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오.
print(round(pv, 4))
if pv < 0.05:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```

##### 마무리 문제 3번
주어진 데이터에는 학생 1,000명의 인종과 과목별 시험 성적이 담겨져 있다. <br>
인종(A그룹, B그룹, C그룹, D그룹)에 따라 수학 성적에 차이가 있는지 일원분산분석을 통해 답하고자 한다. <br>
가설은 아래와 같다.

-  $H_0 : \overline{X_A} = \overline{X_B} = \overline{X_C} = \overline{X_D}$
-  $H_1 : {} \sim H_0 (H_0$가 아니다.)

단, 주어진 데이터가 정규 분포와 등분산을 만족한다고 가정한다.

```python
import pandas as pd
df = pd.read_csv('./StudentsPerformance1.csv', encoding='cp949')
df
```

```python
from scipy import stats

# 각 그룹에 대한 표본평균을 구하시오 (반올림하여 소수 둘째자리까지 계산)
group_A = df[df['인종']=='A그룹']['수학 점수']
group_B = df[df['인종']=='B그룹']['수학 점수']
group_C = df[df['인종']=='C그룹']['수학 점수']
group_D = df[df['인종']=='D그룹']['수학 점수']

print(round(group_A.mean(), 2), round(group_B.mean(), 2), round(group_C.mean(), 2), round(group_D.mean(), 2))

# 위의 가설을 검정하기 위한 검정통계량을 구하시오 (반올림하여 소수 둘째자리까지 계산)
st, pv = stats.f_oneway(group_A, group_B, group_C, group_D)
print(round(st, 2))

# 위의 통계량에 대한 p-value 구하고(반올림 소수 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오.
print(round(pv, 4))
if pv < 0.05:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```

##### 마무리 문제 4번 ★
주어진 데이터에는 1,000명의 학생들의 성별과 학업 성적이 담겨져 있다. <br>
성별에 따라 학업 성적(우수, 보통, 낮음)에 차이가 있는지 카이제곱검정을 통해 답하고자 한다. <br>
가설은 아래와 같다.

-  $H_0 : $ 성별과 학업 성적은 관련이 없다.
-  $H_1 : $ 성별과 학업 성적은 관련이 있다.

```python
import pandas as pd
df = pd.read_csv('./StudentsPerformance2.csv', encoding='cp949')
df
```

```python
from scipy import stats

# 주어진 데이터로 성별과 학업 성적 간의 분할표를 만들었을 때, 학업 성적이 우수한 남학생 수와 여학생 수의 차이를 정수로 출력하시오.
table = pd.crosstab(df['성별'], df['학업 성적'])
print(table)

X1 = table.loc['남성', :]
X2 = table.loc['여성', :]

print(int(abs(X1['우수'] - X2['우수'])))

# 위의 가설을 검정하기 위한 검정통계량을 구하시오 (반올림하여 소수 둘째자리까지 계산)
result = stats.chi2_contingency([X1, X2])
print(round(result.statistic, 2))

# 위의 통계량에 대한 p-value 구하고(반올림 소수 넷째자리까지 계산), 유의수준 0.05하에서 가설검정의 결과를 채택/기각 중 하나로 선택하시오.
print(round(result.pvalue, 4))
if pv < 0.05:
    print("귀무가설 기각")
else:
    print("귀무가설 채택")
```
