```python
X_train = pd.read_csv('/kaggle/input/big-data-analytics-certification/t2-1-train.csv')
X_test = pd.read_csv('/kaggle/input/big-data-analytics-certification/t2-1-test.csv')
```

```python
y_train = X_train['TravelInsurance']
X_train = X_train.drop(columns='TravelInsurance')
```

```python
X_train.isnull().sum()
```

```python
X_test.isnull().sum()
```

```python
X_train_drop = X_train.drop(columns=['id', 'Employment Type'])
X_test_drop = X_test.drop(columns=['id', 'Employment Type'])
```

```python
X_train_drop = X_train_drop.fillna(X_train['AnnualIncome'].mean())
X_test_drop = X_test_drop.fillna(X_test['AnnualIncome'].mean())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

n_col = ['Age', 'AnnualIncome', 'FamilyMembers', 'ChronicDiseases']
v_col = ['GraduateOrNot', 'FrequentFlyer', 'EverTravelledAbroad']

ss = StandardScaler()
ss.fit_transform(X_train_drop[n_col])
ss.transform(X_test_drop[n_col])

le = LabelEncoder()
for i in v_col:
    X_train_drop[i] = le.fit_transform(X_train_drop[i])
    X_test_drop[i] = le.transform(X_test_drop[i])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

X_tr, X_val, y_tr, y_val = train_test_split(X_train_drop, y_train, test_size=0.2, random_state=41)

rf = RandomForestClassifier()
rf.fit(X_tr, y_tr)

print(roc_auc_score(y_tr, rf.predict_proba(X_tr)[:,1]))
print(roc_auc_score(y_val, rf.predict_proba(X_val)[:,1]))
```

```python
pd.DataFrame({'id':X_test.id, 'TravelInsurance':rf.predict_proba(X_test_drop)[:,1]}).to_csv('000000000.csv', index=False)
```
