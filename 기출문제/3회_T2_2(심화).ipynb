```python
X_train = pd.read_csv('/kaggle/input/big-data-analytics-certification/t2-2-X_train.csv')
X_test = pd.read_csv('/kaggle/input/big-data-analytics-certification/t2-2-X_test.csv')
y_train= pd.read_csv('/kaggle/input/big-data-analytics-certification/t2-2-y_train.csv')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score

ss = StandardScaler()
ss.fit_transform(X_train)
ss.transform(X_test)
y = y_train.drop(columns='Serial No.')

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y, test_size=0.2, random_state=41)

rf = RandomForestRegressor()
xgb = XGBRegressor()

rf.fit(X_tr, y_tr)
rf_pred = rf.predict(X_tr)
print(r2_score(y_tr, rf_pred))
print(r2_score(y_val, rf.predict(X_val)))

xgb.fit(X_tr, y_tr)
xgb_pred = xgb.predict(X_tr)
print(r2_score(y_tr, xgb_pred))
print(r2_score(y_val, xgb.predict(X_val)))
```

```python
pd.DataFrame({'Serial No.': X_test['Serial No.'], 'Chance of Admit': xgb.predict(X_test)}).to_csv('000000000.csv', index=False)
```
