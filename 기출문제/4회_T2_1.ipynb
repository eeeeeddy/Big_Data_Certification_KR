```python
X_train = pd.read_csv('/kaggle/input/big-data-analytics-certification-kr-2022/train.csv')
X_test = pd.read_csv('/kaggle/input/big-data-analytics-certification-kr-2022/test.csv')
```

```python
y_train = X_train['Segmentation']
X_train_drop = X_train.drop(columns=['ID', 'Segmentation'])
X_test_drop = X_test.drop(columns='ID')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

ss = StandardScaler()
le = LabelEncoder()

n_col = ['Age', 'Work_Experience', 'Family_Size']
v_col = ['Gender', 'Ever_Married', 'Graduated', 'Profession', 'Spending_Score', 'Var_1']

ss.fit_transform(X_train_drop[n_col])
ss.transform(X_test_drop[n_col])

for i in v_col:
    X_train_drop[i] = le.fit_transform(X_train_drop[i])
    X_test_drop[i] = le.transform(X_test_drop[i])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import f1_score

X_tr, X_val, y_tr, y_val = train_test_split(X_train_drop, y_train, test_size=0.2, random_state=41, stratify=y_train)

rf = RandomForestClassifier()
#xgb = XGBClassifier()

rf.fit(X_tr, y_tr)
print(f1_score(y_tr, rf.predict(X_tr), average='macro'))
print(f1_score(y_val, rf.predict(X_val), average='macro'))
```

```python
pd.DataFrame({"ID":X_test.ID, 'Segmentation':rf.predict(X_test_drop)}).to_csv('000000000.csv', index=False)
```
