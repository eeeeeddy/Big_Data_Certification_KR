57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 
- Unknown : N
- Less than $40K : a
- $40K - $60K : b
- $60K - $80K : c
- $80K - $120K : d
- $120K + : e

```python
df['newIncome'] = df['Income_Category'].map({"Unknown":"N",
                                             "Less than $40K" : "a",
                                             "$40K - $60K" : "b",
                                             "$60K - $80K" : "c",
                                             "$80K - $120K" : "d", 
                                             "$120K +" : "e"})

df.head()
```

58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 
- Unknown : N
- Less than $40K : a
- $40K - $60K : b
- $60K - $80K : c
- $80K - $120K : d
- $120K + : e

```python
def income(x):
    if x == "Unknown":
        return "N"
    elif x == "Less than $40K":
        return "a"
    elif x == "$40K - $60K":
        return "b"
    elif x == "$60K - $80K":
        return "c"
    elif x == "$80K - $120K":
        return "d"
    elif x == "$120K +":
        return "e"
    
    
df['newIncome'] = df.Income_Category.apply(lambda x:income(x))
df['newIncome'].head()
```

59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. 
- (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라

```python
def age(x):
    if x>=0 and x<10:
        return 0
    elif x>=10 and x<20:
        return 10
    elif x>=20 and x<30:
        return 20
    elif x>=30 and x<40:
        return 30
    elif x>=40 and x<50:
        return 40
    elif x>=50 and x<60:
        return 50
    elif x>=60 and x<70:
        return 60
    elif x>=70 and x<80:
        return 70

df['AgeState'] = df["Customer_Age"].apply(lambda x : age(x))
df['AgeState'].value_counts().sort_index()
```

67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. <br>
    해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라 ★

```python
def func(x):
    import datetime
    
    if x.year > 2060:
        return pd.to_datetime(datetime.date(x.year - 100, x.month, x.day))
    else:
        return pd.to_datetime(datetime.date(x.year, x.month, x.day))


df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(func)
df.head()
```

#### 07_Merge , Concat

```python
# 국가별 5세이하 사망비율 통계 데이터
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)

df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

91. df1과 df2 데이터를 하나의 데이터 프레임으로 합쳐라

```python
pd.concat([df1,df2])
```

92. df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 둘다 포함하고 있는 년도에 대해서만 고려한다

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
pd.concat([df3, df4], join='inner')
```

93. df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 모든 컬럼을 포함하고, 결측치는 0으로 대체한다

```python
pd.concat([df3, df4], join='outer').fillna(0)
```
