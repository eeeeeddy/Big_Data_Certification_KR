# **DataFrame**

```python
import pandas as pd
import numpy as np
```

```python
pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]])
```

```python
pd.DataFrame([[1,2,3],
              [4,5,6],
              [7,8,9]], columns=['가','나','다'])
```

```python
data = {
    'name': ['Kim', 'Lee', 'Park'], 
    'age': [24, 27, 34], 
    'children': [2, 1, 3]
}

df = pd.DataFrame(data)
df
```

```python
# 2개 이상의 column 선택은 fancy indexing으로 가능하다.

df[['name','children']]
```

```python
df.rename(columns={'name':'이름'})
```

```python
# inplace=True 옵션으로 변경사항을 바로 적용할 수 있습니다.

df.rename(columns={'name':'이름'}, inplace=True)
df
```

연습문제

```python
df = pd.DataFrame([['KFC', 1000, 4.5],
                   ['McDonald', 2000, 3.9],
                   ['SchoolFood', 2500, 4.2]], columns=['food', 'price', 'rating'])
df
```

```python
df[['food','rating']]
```

```python
# rename 함수 안에 작성법 유의하기
df.rename(columns = {"food": "place"})
```
