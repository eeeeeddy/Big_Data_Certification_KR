#### 서비스 이탈예측 데이터
고객의 신상정보 데이터를 통한 회사 서비스 이탈 예측 (종속변수 : Exited)

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
drop_col = ['CustomerId', 'Surname']
x_train = x_train.drop(columns=drop_col)
x_test = x_test.drop(columns=drop_col)
display(x_train, x_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# 데이터 더미화
x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)

# train 데이터와 컬럼 순서 동일하게 하기
x_test = x_test[x_train.columns]
y = y_train['Exited']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_train, y, test_size=0.33, random_state=42)
```

```python
rf = RandomForestClassifier(random_state=23)
rf.fit(X_train, y_train) # X_train, y_train으로 훈련
```

```python
rf.predict(X_train)
rf.predict_proba(X_train)[:,1]

rf.predict(X_test) # >> X_test 데이터를 이용하여 모델 훈련 및 검증
rf.predict_proba(X_test)[:,1]
```

predict는 각각의 모델의 최종적인 예측값을 출력하는 결과.
- 회귀 모델의 경우 : 0.98733..과 같이 구체적으로 특정 값을 출력
- 분류 모델의 경우 : 0,1,2와 같이 가장 확률이 높은 클래스를 출력

이때, 분류 모델이 각 클래스에 대한 확률을 일일이 출력하는 것이 predict_proba <br>
(predict_proba는 predict probability, 즉 각 클래스의 확률을 예측하라는 의미.)

따라서 회귀모델에는 predict_proba를 사용할 수 없으며, 출력되는 값은 다음과 같다.

- 클래스가 3개인 문제를 풀 경우 <br>
Input > Output : [클래스가 0일 확률, 클래스가 1일 확률, 클래스가 2일 확률]

predict_proba[:,1]의 방식은 이진분류 문제일 때, 회귀 문제를 분류 모델을 통해 푸는 방식을 의미.

```python
# 모델 성능 확인하기
from sklearn.metrics import accuracy_score, f1_score, recall_score, roc_auc_score, precision_score

# accuracy score
print('훈련 정확도 : ', accuracy_score(y_train, rf.predict(X_train)))
print('검증 정확도 : ', accuracy_score(y_test, rf.predict(X_test)))
print('\n')
# f1_score
print('훈련 정확도 : ', f1_score(y_train, rf.predict(X_train)))
print('검증 정확도 : ', f1_score(y_test, rf.predict(X_test)))
print('\n')
# recall_score
print('훈련 정확도 : ', recall_score(y_train, rf.predict(X_train)))
print('검증 정확도 : ', recall_score(y_test, rf.predict(X_test)))
print('\n')
# precision_score
print('훈련 정확도 : ', precision_score(y_train, rf.predict(X_train)))
print('검증 정확도 : ', precision_score(y_test, rf.predict(X_test)))
print('\n')
# roc_auc_score
print('훈련 정확도 : ', roc_auc_score(y_train, rf.predict_proba(X_train)[:,1]))
print('검증 정확도 : ', roc_auc_score(y_test, rf.predict_proba(X_test)[:,1]))
print('\n')
```

```python
# test 데이터 예측
rf.predict(x_test)
rf.predict_proba(x_test)[:,1]
```

```python
# 답안 제출

pd.DataFrame({'CustomerId' : x_test.CustomerId, 'Exited' : rf.predict(x_test)}).to_csv('000000000.csv', index=False)

pd.DataFrame({'CustomerId' : x_test.CustomerId, 'Exited' : rf.predict_proba(x_test)[:,1]}).to_csv('000000000.csv', index=False)
```
